очень много пропусков, в целом все выглядит плохо.

```python
df['balcony'] = df['balcony'].fillna(0)
```

избавился от пропусков в балконах. Самое просто

```python
pd.to_numeric(df['ceiling_height'], errors='coerce')
ceiling_height_mean = df['ceiling_height'].mean()
df['ceiling_height'].fillna(ceiling_height_mean, inplace=True)
df.head()

```

Добавил среднюю высоту потолков

```python
def fillna_ceiling_height(row):
    if row['cityCenters_nearest'] < 2500:
        return 'ДС'
    elif row['cityCenters_nearest'] < 5000:
        return 'ДС1'
    elif row['cityCenters_nearest'] < 17000:
        return 'ДС2'
    elif row['cityCenters_nearest'] >= 17000:
        return 'ДС3'
df['fromcenter_category'] = df.apply(fillna_ceiling_height, axis=1)
df[['locality_name', 'cityCenters_nearest', 'fromcenter_category']].head(10)
```

Но если заморочиться и разбить районы по отдаленности от центра, а потом выдать им средние значения по потолками, то будет куда объективнее

```python
df.groupby('fromcenter_category')['ceiling_height'].mean().sort_values(ascending=False).to_frame()
list_of_fromcenter_cat = ['ДС', 'ДС1', 'ДС2', 'ДС3']
for column in list_of_fromcenter_cat:
    value_for_filling = df.groupby('fromcenter_category')['ceiling_height'].mean().get(key=column)
    df['ceiling_height'] = df.groupby('fromcenter_category')['ceiling_height'].apply(lambda x: x.fillna(x.mean()))
df['ceiling_height'] = df.groupby('fromcenter_category')['ceiling_height'].apply(lambda x: x.fillna(x.mean()))
df.head()
```

Красиво

```python
print('Количество NaN в столбце is_apartment:', df['is_apartment'].isnull().sum())
df['is_apartment'].value_counts()
```

Скорее всего люди просто не знают или не считают нужным заполнять эту строку, но просто поменять значения False нельзя, так как тогда выйдет всего 50 строк на 21тысячу в данных

```python
df['sqmtr'] = df[df['last_price'].notnull()]['last_price']/((df['total_area']))
df.head()
```

```python
pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df['weekday'] = pd.to_datetime(df['first_day_exposition']).dt.weekday
dict_of_weekdays = {1: '1 - понедельник', 2: '2 - вторник', 3: '3 - среда', 4: '4 - четверг', 
                    5: '5 - пятница', 6: '6 - суббота', 7: '7 - воскресенье'}
df['weekday'] = df['weekday'].map(dict_of_weekdays)

```

Добавил колонки цены средней стоимости и нумерацию дня недели. Как ни старался-добавить буквенные значения дням недели не удалось.

```python
def determine_level_of_floor(row):
    floor = row['floor']
    total_floors = row['floors_total']

    if floor == 1:
        return 'первый'
    elif floor == total_floors:
        return 'последний'
    else:
        return 'другой'

df['floor_level'] = df.apply(determine_level_of_floor, axis=1)
```

```python
df['liv_to_full_ratio'] = df[df['total_area'].notnull()]['living_area']/((df['total_area']))
df['kit_to_full_ratio'] = df[df['total_area'].notnull()]['kitchen_area']/((df['total_area']))
df.head(10)
```

liv_to_full_ratio соотношение жилплощади к общей; kit_to_full_ratio Кухня к общей

```python
list_to_int = ['last_price', 'floors_total', 'balcony', 'airports_nearest', 'cityCenters_nearest', 'parks_around3000',
               'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition', 'sqmtr']
for column in list_to_int:
    df[column] = df[df[column].notnull()][column].astype('int')
```

перевел в инт

```python
list_to_float_1 = ['total_area', 'ceiling_height', 'living_area', 'kitchen_area']
for column in list_to_float_1:
    df[column] = df[df[column].notnull()][column].round(1)
```

```python
list_to_float_1 = ['liv_to_full_ratio', 'kit_to_full_ratio']
for column in list_to_float_1:
    df[column] = df[df[column].notnull()][column].round(2)
df.head(10)
```
